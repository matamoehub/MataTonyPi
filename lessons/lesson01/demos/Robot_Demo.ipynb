# TonyPi Show Time

This demo is meant to feel like a mini robot performance, not just an API checklist.

## Performance Plan

TonyPi will:
1. Enter with a greeting
2. Show expressive head and arm actions
3. Walk, turn, and sidestep
4. Demonstrate vision calls
5. Pretend to pick up an object
6. Finish with a celebration

```python
from lesson_loader import setup
setup(verbose=True)

from student_robot_v2 import bot
import time

myRobot = bot()
myRobot.head.center()
time.sleep(0.5)
```

```python
def stage(title):
    print(f"\n=== {title} ===")


def pause(seconds=0.35):
    time.sleep(seconds)
```

```python
stage("Grand entrance")
myRobot.head.center()
pause(0.5)
myRobot.say("Hello everyone. TonyPi is ready to perform.")
pause(0.4)
myRobot.anim.greet()
pause(0.4)
myRobot.anim.wave()
pause(0.4)
myRobot.pose.ready()
pause(0.5)
```

```python
stage("Head and arms")
myRobot.head.center()
pause(0.4)
myRobot.head.look_left()
pause()
myRobot.head.look_right()
pause()
myRobot.head.look_up()
pause()
myRobot.head.look_down()
pause()
myRobot.head.center()
pause(0.4)
myRobot.head.nod()
pause(0.4)
myRobot.head.shake()
pause(0.4)
myRobot.arms.left_up()
pause(0.4)
myRobot.arms.right_up()
pause(0.4)
myRobot.arms.hands_up()
pause(0.4)
myRobot.arms.open()
pause(0.4)
myRobot.arms.close()
pause(0.4)
myRobot.arms.center()
pause(0.5)
```

```python
stage("Movement and poses")
myRobot.head.center()
pause(0.4)
myRobot.pose.bow()
pause(0.5)
myRobot.stand()
pause(0.4)
myRobot.motion.walk_forward(steps=1)
pause(0.35)
myRobot.motion.walk_backward(steps=1)
pause(0.35)
myRobot.motion.turn_left(steps=1)
pause(0.35)
myRobot.motion.turn_right(steps=1)
pause(0.35)
myRobot.motion.step_left(steps=1)
pause(0.35)
myRobot.motion.step_right(steps=1)
pause(0.35)
myRobot.anim.dance()
pause(0.5)
myRobot.motion.stop()
pause(0.5)
```

```python
stage("Vision showcase")
myRobot.head.center()
pause(0.6)
print(myRobot.vision.find_color("blue").to_dict())
pause(0.25)
print(myRobot.vision.find_object("cube").to_dict())
pause(0.25)
print(myRobot.vision.find_face().to_dict())
pause(0.25)
print(myRobot.vision.find_tag(7).to_dict())
pause(0.3)
myRobot.vision.track_color("blue")
pause(0.25)
myRobot.vision.track_face()
pause(0.25)
myRobot.vision.scan_for("cube")
pause(0.25)
myRobot.vision.snapshot()
pause(0.5)
```

```python
stage("Pickup story")
myRobot.head.center()
pause(0.4)
myRobot.say("I found a cube. I will bring it over.")
pause(0.4)
myRobot.pickup.approach_object("cube")
pause(0.4)
myRobot.pickup.pick_up("cube")
pause(0.4)
myRobot.pickup.carry()
pause(0.4)
myRobot.pickup.place_down()
pause(0.4)
myRobot.pickup.release()
pause(0.5)
```

```python
stage("Finale")
myRobot.head.center()
pause(0.4)
myRobot.say("Thank you for watching my TonyPi show.")
pause(0.4)
myRobot.anim.think()
pause(0.4)
myRobot.anim.yes()
pause(0.4)
myRobot.anim.celebrate()
pause(0.4)
myRobot.voice.celebrate()
pause(0.4)
myRobot.home()
```
