# Lesson 1 - TonyPi Feature Tour

In this lesson, students meet TonyPi through the `student_robot_v2` library and try each major feature one step at a time.

## Step 1: Connect to the lesson robot

Run this cell first. It loads the lesson helper and creates `myRobot`.

```python
from lesson_header import *

myRobot.status()
```

## Step 2: Make TonyPi talk and greet

TonyPi can speak and play friendly animations.

```python
myRobot.say("Hello everyone. I am TonyPi.")
myRobot.anim.wave()
myRobot.anim.greet()
```

## Step 3: Move the head and arms

These commands help students learn that the robot is organized into clear namespaces.

```python
myRobot.head.look_left()
myRobot.head.look_right()
myRobot.head.look_up()
myRobot.head.center()

myRobot.arms.left_up()
myRobot.arms.right_up()
myRobot.arms.hands_up()
myRobot.arms.center()
```

## Step 4: Try poses and simple movement

These actions show how TonyPi can change body position and move around.

```python
myRobot.pose.ready()
myRobot.stand()
myRobot.motion.walk_forward(steps=1)
myRobot.motion.turn_left(steps=1)
myRobot.motion.step_right(steps=1)
myRobot.motion.stop()
myRobot.sit()
```

## Step 5: Explore vision features

The current classroom build uses stub vision results, but the lesson still shows students how the API works.

```python
color_result = myRobot.vision.find_color("red")
face_result = myRobot.vision.find_face()
tag_result = myRobot.vision.find_tag(1)

print(color_result.to_dict())
print(face_result.to_dict())
print(tag_result.to_dict())

myRobot.vision.scan_for("ball")
myRobot.vision.snapshot()
```

## Step 6: Practice pickup actions

These commands introduce the pickup workflow even if the backend is still mocked.

```python
myRobot.pickup.approach_object("block")
myRobot.pickup.grab()
myRobot.pickup.carry()
myRobot.pickup.place_down()
myRobot.pickup.release()
```

## Step 7: Build a short robot show

This final cell combines speech, expression, and movement into one simple routine students can remix.

```python
myRobot.say("Watch my robot show.")
myRobot.anim.dance()
myRobot.head.scan()
myRobot.motion.turn_right(steps=1)
myRobot.anim.celebrate()
myRobot.voice.celebrate()
myRobot.home()
```
