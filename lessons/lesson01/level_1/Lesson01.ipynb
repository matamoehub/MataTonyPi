## Feature Goals

In this lesson you will try:

## Step 1: Load the lesson robot

Run this first so `myRobot` is ready to use.

```python
from lesson_header import *

print(myRobot.status())
```

## Step 2: Hello TonyPi

Start with a simple greeting. Change the words if you want.

```python
myRobot.say("Hello everyone. I am TonyPi.")
myRobot.anim.wave()
myRobot.anim.greet()
```

## Step 3: Head and arms

TonyPi can look around and use expressive arm poses.

```python
myRobot.head.look_left()
myRobot.head.look_right()
myRobot.head.look_up()
myRobot.head.look_down()
myRobot.head.center()

myRobot.arms.left_up()
myRobot.arms.right_up()
myRobot.arms.hands_up()
myRobot.arms.open()
myRobot.arms.close()
myRobot.arms.center()
```

## Step 4: Poses and movement

TonyPi can change poses, walk, turn, sidestep, and stop.

```python
myRobot.pose.ready()
myRobot.pose.bow()
myRobot.stand()
myRobot.motion.walk_forward(steps=1)
myRobot.motion.walk_backward(steps=1)
myRobot.motion.turn_left(steps=1)
myRobot.motion.turn_right(steps=1)
myRobot.motion.step_left(steps=1)
myRobot.motion.step_right(steps=1)
myRobot.motion.stop()
myRobot.sit()
```

## Step 5: Vision explorer

The current TonyPi classroom build returns stub results, but this still shows how students will use the vision API.

```python
color_result = myRobot.vision.find_color("red")
object_result = myRobot.vision.find_object("ball")
face_result = myRobot.vision.find_face()
tag_result = myRobot.vision.find_tag(1)

print(color_result.to_dict())
print(object_result.to_dict())
print(face_result.to_dict())
print(tag_result.to_dict())

myRobot.vision.track_color("red")
myRobot.vision.track_face()
myRobot.vision.scan_for("ball")
myRobot.vision.snapshot()
```

## Step 6: Pickup routine

These commands introduce the pickup flow that later lessons can build on.

```python
myRobot.pickup.approach_object("block")
myRobot.pickup.pick_up("block")
myRobot.pickup.grab()
myRobot.pickup.carry()
myRobot.pickup.place_down()
myRobot.pickup.release()
```

## Performance Plan

Build a mini show where TonyPi:
1. Greets the audience
2. Looks around the room
3. Walks and turns with style
4. Pretends to find an object
5. Finishes with a celebration

```python
myRobot.say("Watch my TonyPi show.")
myRobot.anim.greet()
myRobot.head.scan()
myRobot.motion.walk_forward(steps=1)
myRobot.motion.turn_right(steps=1)
myRobot.vision.scan_for("cube")
myRobot.anim.think()
myRobot.pickup.approach_object("cube")
myRobot.anim.celebrate()
myRobot.voice.celebrate()
myRobot.home()
```
