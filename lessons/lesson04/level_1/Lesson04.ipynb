## Activity 2


```python
result = myRobot.vision.find_color("red")
print(result.to_dict())
```

## Activity 3


```python
if result.found:
    myRobot.say("I found red!")
    myRobot.anim.yes()
else:
    myRobot.say("I do not see red yet.")
    myRobot.anim.scan()
```
