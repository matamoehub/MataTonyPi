## Activity 1


```python
for color in ["red", "green", "blue", "yellow"]:
    seen = myRobot.vision.find_color(color)
    print(color, seen.to_dict())
```

## Activity 2


```python
detected = myRobot.vision.find_color("blue")
if detected.found:
    myRobot.say("Blue means dance time!")
    myRobot.anim.dance()
else:
    myRobot.say("No blue found. I will keep scanning.")
    myRobot.anim.scan()
```
