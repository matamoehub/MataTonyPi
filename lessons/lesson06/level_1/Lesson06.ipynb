# Lesson 6 - Target Position Recognition

Use color detection coordinates to decide if a target is left, center, or right.


## Setup
Run the next cell first to initialize `myRobot`.


```python
from lesson_header import *
print(myRobot.status())
```

## Activity 1


```python
target = myRobot.vision.find_color("green")
print(target.to_dict())
```

## Activity 2


```python
if target.found:
    if target.x < 250:
        myRobot.say("Target is on my left")
        myRobot.head.look_left()
    elif target.x > 390:
        myRobot.say("Target is on my right")
        myRobot.head.look_right()
    else:
        myRobot.say("Target is centered")
        myRobot.head.center()
else:
    myRobot.say("I cannot locate the target yet")
```
