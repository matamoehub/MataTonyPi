# Lesson 7 - Object Tracking

Continuously track a color target and turn body/head toward it.


## Setup
Run the next cell first to initialize `myRobot`.


```python
from lesson_header import *
print(myRobot.status())
```

## Activity 1


```python
for _ in range(5):
    track = myRobot.vision.track_color("yellow")
    print(track.get("objects", []))
```

## Activity 2


```python
found = myRobot.vision.find_color("yellow")
if found.found:
    if found.x < 260:
        myRobot.motion.turn_left(steps=1)
    elif found.x > 380:
        myRobot.motion.turn_right(steps=1)
    else:
        myRobot.motion.walk_forward(steps=1)
else:
    myRobot.anim.scan()
```
