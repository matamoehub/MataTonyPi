## Activity 1


```python
ball = myRobot.vision.find_color("red")
print(ball.to_dict())
```

## Activity 2


```python
if ball.found:
    myRobot.say("Approaching the ball")
    myRobot.motion.walk_forward(steps=1)
    myRobot.anim.run("left_kick")
else:
    myRobot.say("Find the ball first")
    myRobot.anim.scan()
```
