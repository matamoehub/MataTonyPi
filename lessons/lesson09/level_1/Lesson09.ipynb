## Activity 2


```python
if line.found:
    if line.x < 250:
        myRobot.motion.step_left(steps=1)
    elif line.x > 390:
        myRobot.motion.step_right(steps=1)
    else:
        myRobot.motion.walk_forward(steps=1)
else:
    myRobot.motion.stop()
    myRobot.say("Line lost")
```
