# Lesson 13 - Rock Paper Scissors

In this lesson, TonyPi uses hand recognition and expressive gestures to play rock paper scissors with you.

## What We Will Do

1. Get TonyPi ready
2. Count down together
3. Capture the player hand sign
4. Let TonyPi choose a move
5. Compare the moves
6. Celebrate the result


```
from lesson_header import *
import random
import time

print(myRobot.status())
```

## Step 1: Get ready


```
myRobot.head.center()
myRobot.stand()
myRobot.arms.open()
myRobot.say("Let us play rock paper scissors.")
```

## Step 2: TonyPi chooses a move


```
def show_robot_move(move):
    if move == "rock":
        myRobot.arms.close()
    elif move == "paper":
        myRobot.arms.open()
    else:
        myRobot.arms.left_up()
        myRobot.arms.right_up()

def throw_motion():
    myRobot.arms.hands_up()
    time.sleep(0.12)
    myRobot.arms.center()
    time.sleep(0.10)

def countdown():
    beats = ["Ready", "Rock", "Paper", "Scissors", "Go"]
    for word in beats:
        myRobot.say(word)
        throw_motion()
        time.sleep(0.15)
```

## Step 3: Count down and capture the player hand


```
countdown()
robot_move = random.choice(["rock", "paper", "scissors"])
hand_result = myRobot.vision.recognize_hands(show=True)
print(hand_result)

possible_moves = hand_result.get("game_moves", [])
player_move = possible_moves[0] if possible_moves else None
print("Player move:", player_move)
```

## Step 4: Show TonyPi's move and compare


```
show_robot_move(robot_move)
print("Robot move:", robot_move)

def winner(player, robot):
    if player is None:
        return "retry"
    if player == robot:
        return "draw"
    wins = {("rock", "scissors"), ("paper", "rock"), ("scissors", "paper")}
    return "player" if (player, robot) in wins else "robot"

result = winner(player_move, robot_move)
print("Result:", result)
```

## Step 5: React to the result


```
if result == "retry":
    myRobot.say("I could not read your hand. Try again.")
    myRobot.anim.think()
elif result == "draw":
    myRobot.say("It is a draw.")
    throw_motion()
elif result == "player":
    myRobot.say("You win this round.")
    myRobot.pose.bow()
else:
    myRobot.say("I win this round.")
    myRobot.anim.celebrate()

myRobot.home()
```
