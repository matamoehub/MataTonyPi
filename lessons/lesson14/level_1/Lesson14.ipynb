# Lesson 14 - Multi-Control Robot Team

In this lesson, you will set up a group of TonyPi robots so one robot acts as the server and the others act as clients. Then you will test synchronized action-group moves.

## What We Will Do

1. Choose the server robot in robot-console
2. Configure the client robots
3. Restart the TonyPi control runtime
4. Test synchronized action groups
5. Build a short team performance


## Step 1: Configure the robot team in robot-console

Use the new TonyPi multi-control API in robot-console:

- open the live TonyPi list
- choose one robot as the server
- choose the other robots as clients
- apply the configuration

The console uses these endpoints:

- `GET /api/tonypi/multi-control/live`
- `POST /api/tonypi/multi-control/configure`


## Step 2: Restart the TonyPi runtime

After applying the server/client configuration, restart the TonyPi runtime or reboot the robots before testing. This matches the upstream TonyPi group-control setup flow.


## Step 3: Connect to the server robot notebook

Open a notebook on the server robot and load TonyPi. The server robot will forward action commands to the clients through the upstream multi-control scripts.


```
from lesson_header import *
import time

print(myRobot.status())
```

## Step 4: Test basic synchronized actions

Start with simple action groups that are easy to spot across the whole robot team.


```
myRobot.anim.run("stand")
time.sleep(1.0)
myRobot.anim.run("wave")
time.sleep(1.0)
myRobot.anim.run("bow")
time.sleep(1.0)
myRobot.anim.run("twist")
```

## Step 5: Inspect the action-group catalog

The teaching layer can now show all installed TonyPi action groups, including the dance range.


```
print(myRobot.anim.catalog())
print(myRobot.anim.dance_moves())
myRobot.anim.show_action_groups()
```

## Step 6: Try a dance move by id

On this TonyPi setup, ids `16-24` are dance moves.


```
myRobot.anim.run_id(16)
time.sleep(1.0)
myRobot.anim.run_id(17)
```

## Step 7: Build a short team routine

Use a mix of greeting, pose, and dance actions to make the robot group perform together.


```
myRobot.say("TonyPi team ready.")
myRobot.anim.run("stand")
time.sleep(0.8)
myRobot.anim.run("wave")
time.sleep(0.8)
myRobot.anim.run_id(16)
time.sleep(0.8)
myRobot.anim.run("bow")
```
