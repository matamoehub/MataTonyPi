# Lesson 15 - Wireless Joystick Control

In this lesson, you will learn how the TonyPi wireless joystick is mapped to robot actions.

## What We Will Do

1. Load TonyPi
2. Inspect the controller button map
3. Inspect the dance button shortcuts
4. Review the controller modes
5. Match joystick buttons to robot API actions


```
from lesson_header import *

print(myRobot.status())
```

## Step 1: Show the controller summary


```
print(myRobot.controller.summary())
```

## Step 2: Show the normal button map


```
myRobot.controller.show_buttons()
```

## Step 3: Show the dance shortcuts

The dance buttons match the upstream TonyPi joystick combinations.


```
myRobot.controller.show_dance_buttons()
```

## Step 4: Controller modes


```
print(myRobot.controller.modes())
```

## Step 5: Match joystick buttons to TonyPi actions

These examples show how some controller buttons relate to the student API.


```
# Triangle on the controller is a wave
myRobot.anim.wave()

# Cross on the controller is a bow
myRobot.pose.bow()

# Arrow directions map to movement
myRobot.motion.walk_forward(steps=1)
myRobot.motion.walk_backward(steps=1)
myRobot.motion.step_left(steps=1)
myRobot.motion.step_right(steps=1)
```

## Step 6: Try a dance shortcut by id

The controller dance buttons correspond to the TonyPi dance action-group range.


```
print(myRobot.anim.dance_moves())
myRobot.anim.run_id(16)
```
